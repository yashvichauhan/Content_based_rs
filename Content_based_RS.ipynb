```
movies_df['year'] = movies_df['title'].str.extract('(\(\d\d\d\d\))',expand=False)
```

```
movies_df['year']=movies_df['year'].str.extract('(\d\d\d\d)',expand=False)
```

```
movies_df['title']=movies_df['title'].str.replace('(\(\d\d\d\d\))','')
```

```
movies_df['title']=movies_df['title'].str.strip()
```

```
movies_df['year'].isna().value_counts()
```

```
movies_df['genres']=movies_df['genres'].apply(lambda string:string.split('|'))
```

```
movies_df
```

```
genres_df=movies_df.copy()
```

```
for indx,row in genres_df.iterrows():
  for x in row['genres']:
    genres_df.at[indx,x]=1
```

```
genres_df
```

```
genres_df.fillna(0,inplace=True)
```

```
rating_df.isna().any()
```

```
rating_df.drop('timestamp',axis=1,inplace=True)
```

```
rating_df['rating'].unique()
```

```
#input user
input_df=pd.DataFrame(
    [['Toy Story',2.2],
     ['Jumanji',3.5],
     ['Fast and the Furious: Tokyo Drift, The (Fast and the Furious 3, The)',4.1],
     ['Final Destination 5',4.3],
     ['Inception',4.5],
     ['Twilight Saga: Breaking Dawn - Part 2, The',3.2]],columns=['title','rating']
)
```

```
input_df
```

```
input_df=input_df.merge(movies_df,left_on='title',right_on='title')
```

```
input_df
```

```
input_df.drop(['genres','year'],axis=1,inplace=True)
```

```
input_df
```

```
movie_mat = pd.DataFrame(genres_df[genres_df['title'].isin(input_df['title'])])
```

```
movie_mat.drop(['title','genres','year'],axis=1,inplace=True)
```

```
movie_mat.reset_index(0,inplace=True)
```

```
movie_mat.drop(['index','movieId'],axis=1,inplace=True)
```

```
movie_mat
```

```
movie_mat.T
```

```
user_prof=pd.DataFrame(movie_mat.T.dot(input_df['rating']),columns=['agg_r'])
```

```
user_prof
```

```
genre_table = genres_df.set_index(keys=['movieId'])
```

```
genre_table.drop(['title','genres','year'],axis=1,inplace=True)
```

```
genre_table
```

```
user_prof.index
```

```
user_prof1=pd.Series(user_prof['agg_r'],index=user_prof.index)
```

```
type(user_prof1)
```

```
rec_table=genre_table.mul(user_prof['agg_r'],axis=0)
```

```
rec_table= ((genre_table*user_prof1).sum(axis=1)/user_prof1.sum())
```

```
rec_table.sort_values(axis=0,ascending=False,inplace=True)
```
